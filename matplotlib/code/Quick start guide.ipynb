```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
date_list = [1,2,3,4]
stock1 = [4,8,2,6]
stock2 = [10,12,5,3]

plt.plot(date_list, stock1)
plt.plot(date_list, stock2)
plt.show()
```

```python
date_list = [1,2,3,4]
stock1 = [4,8,2,6]
stock2 = [10,12,5,3]

plt.plot(date_list, stock1, "ro--", label = ':abd')
plt.plot(date_list, stock2, "b^--", label = ':def')

plt.title("")
plt.xlabel("")
plt.ylabel("")

#xy
plt.xticks([1,2,3,4])
plt.yticks(np.arange(2,13))

plt.legend()

#
plt.grid()

plt.xlim(2.5, 4.5)
plt.ylim(1.5, 6.5)

plt.show()
```
